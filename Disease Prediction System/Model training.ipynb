```python
import numpy as np
import pandas as pd
from prettytable import PrettyTable
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pickle
```

# Training Data

```python
df = pd.read_csv("data/training.csv", usecols = [i for i in range(133)])
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.columns
```

```python
t = PrettyTable(['Column', 'Nwgative', 'Positive'])
single_value_col = []

for col in df.columns[:132]:    
    try:
        t.add_row([col, df[col].value_counts()[0], df[col].value_counts()[1]])
    except:
        single_value_col.append(col)
```

```python
t
```

```python
single_value_col
```

```python
df.drop(single_value_col, axis=1, inplace= True)
```

```python
df.head()
```

# Testing Data

```python
test = pd.read_csv("data/testing.csv")
```

```python
test.head()
```

```python
test.drop(single_value_col, axis=1, inplace= True)
```

```python
df.shape, test.shape
```

# Data Encoding

```python
x_train = df.drop('prognosis', axis=1)
x_test = test.drop('prognosis', axis=1)
```

```python
le = preprocessing.LabelEncoder()
y_train = le.fit_transform(df['prognosis'])
y_test = le.transform(test['prognosis'])
```

```python
output = open('model/label_enccoder.pkl', 'wb')
pickle.dump(le, output)
output.close()
```

# Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
y_pred
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(confusion_matrix)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
output = open('model/logreg_model.pkl', 'wb')
pickle.dump(logreg, output)
output.close()
```
